### lenguaje Python ,crear la clase "animal" con los atributos "nombre " ,"raza", crear el contructor animal, y las funciones ladrar(), correr(), crear los objetos pitbull, chihuahua,nivel principiante,no es interactivo, proporciona los datos

```python
# Definimos la clase
class Animal:
    # El constructor (__init__) inicializa los atributos del objeto
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    # Función para que el animal ladre
    def ladrar(self):
        print(f"{self.nombre} dice: ¡Guau, guau!🤞")
    # funcion para que el animal muerda

    def Morder(self):
      print(f"{self.nombre} esta mordiendo a la vecina💋")

    # Función para que el animal corra
    def correr(self):
        print(f"{self.nombre}, el {self.raza}, está corriendo velozmente.😍")

# --- Creación de los objetos (Instanciación) ---

# Creamos el objeto pitbull con sus datos correspondientes
pitbull = Animal("Zeus", "Pitbull")

# Creamos el objeto chihuahua con sus datos correspondientes
chihuahua = Animal("Chispa", "Chihuahua")
# creamos el objeto perro callejero con datos
perrocallejero = Animal("Max", "Chafa")

# --- Uso de las funciones ---

# Acciones del Pitbull
print("datos del primer perro")
pitbull.ladrar()
pitbull.correr()

print("-" * 20) # Separador visual

# Acciones del perro callejero
print("datos del segundo perro")
perrocallejero.ladrar()
perrocallejero.correr()
perrocallejero.Morder()

print("-" * 20) # Separador visual

# Acciones del Chihuahua
print("datos del tercer perro")
chihuahua.ladrar()
chihuahua.correr()
print("programa creado por Ramirez Azul NL = 44")
```
